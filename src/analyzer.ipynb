```python
# Function: Convert Pandas DF to dictionary
# Return dictionary
# Show statistics & plots
def df2dict(df):
    dict_2015 = {}
    inter_2015 = {}
    for index, row in df.iterrows():
        if row['id'] not in dict_2015:
            # Add new paper entry
            title = row['title']
            session = row['session']
            author = row['author']
            aff = row['affiliation']
            inst = row['University/Center']
            if not isinstance(inst, float):
                if len(inst.split(',')) > 1:
                    inst = inst.split(',')[1]
                elif len(inst.split('&')) > 1:
                    inst = inst.split('&')[1]
            country = row['country']
            if not isinstance(country, float):
                if len(country.split(',')) > 1:
                    country = country.split(',')[1]
                elif len(country.split('&')) > 1:
                    country = country.split('&')[1]
            industry = row['industry']
            dict_2015[row['id']] = {'title': title, 'session': session, 
                                'author_info': [{'name': author, 'aff': aff, 
                                                'inst': inst, 'country': country, 'industry': industry}]}
        else:
            author = row['author']
            aff = row['affiliation']
            inst = row['University/Center']
            if not isinstance(inst, float):
                if len(inst.split(',')) > 1:
                    inst = inst.split(',')[1]
                elif len(inst.split('&')) > 1:
                    inst = inst.split('&')[1]
            country = row['country']
            if not isinstance(country, float):
                if len(country.split(',')) > 1:
                    country = country.split(',')[1]
                elif  len(country.split('&')) > 1:
                    country = country.split('&')[1]
            industry = row['industry']
            dict_2015[row['id']]['author_info'].append({'name': author, 'aff': aff, 
                                                'inst': inst, 'country': country, 'industry': industry})

    # Stats on Number of Authors 
#     num_author = []
#     for key in dict_2015:
#         num_author.append(len(dict_2015[key]['author_info']))
#     num_author = array(num_author)
#     np.histogram(num_author)
#     print ('Average number of authors per paper:')
#     print (np.average(num_author))
#     print ('Max number of authors per paper:')
#     print (np.amax(num_author))
#     plt.hist(num_author, bins='auto')  # arguments are passed to np.histogram
#     plt.title("Number of Authors per Paper: 2016")
#     plt.show()
    return dict_2015
```

```python
# Function: Compare the field (academia vs industry) of authors for a given paper
# Return a dictionary of papers that have a-i collaborations
def compare_field(dict_data):
    dict_field = {}
    for key in dict_data:
        field_lst = []
        for ele in dict_data[key]['author_info']:
            field_lst.append(ele['industry'])
        if len( set( field_lst ) ) > 1:
            dict_field.update({key: dict_data[key]})
    return dict_field
```

```python
# Function: Compare the countries of authors for a given paper
# Return a dictionary of papers that have international collaborations
def compare_country(dict_data):
    dict_field = {}
    for key in dict_data:
        field_lst = []
        for ele in dict_data[key]['author_info']:
            if not isinstance(ele['country'], float):
                if len(ele['country'].split(',')) > 1:
                    country = ele['country'][0]
                elif  len(ele['country'].split('&')) > 1:
                    country = ele['country'][0]
                else:
                    country = ele['country']
                field_lst.append(country)
            if len( set( field_lst ) ) > 1:
                dict_field.update({key: dict_data[key]})
    return dict_field
```

```python
# Function: Compare the institutes of authors for a given paper
# Return a dictionary of papers that have across-institute collaborations
def compare_institute(dict_data):
    dict_field = {}
    for key in dict_data:
        field_lst = []
        for ele in dict_data[key]['author_info']:
            if not isinstance(ele['inst'], float):
                if len(ele['inst'].split(',')) > 1:
                    field_lst.append(ele['inst'])
                else:
                    field_lst.append(ele['inst'])
        if len( set( field_lst ) ) > 1:
            dict_field.update({key: dict_data[key]})
    return dict_field
```

```python
def paper_per_cty(data_dict):
    cty_paper = {}
    for key in data_dict:
        for ele in data_dict[key]['author_info']:
            cty = ele['country']
            if not isinstance(cty, float):
                if cty not in cty_paper:
                    cty_paper[cty] = 1
                else:
                    cty_paper[cty] += 1
    return cty_paper
```

```python
def create_author_net(dict_2015):
    cty15 = []
    for key in dict_2015:
        cty_lst = []
        for ele in dict_2015[key]['author_info']:
            cty = ele['name']
            cty_lst.append(cty)
        cty15.append(cty_lst)
    
    cty_tp_val = {}
    cty_id = {}
    ct = -1
    for ele in cty15:
        if len(set(ele)) > 1:
            diff_ele = list(set(ele))
            for i in range(len(diff_ele)):
                for j in range(i+1, len(diff_ele)):
                    a1 = diff_ele[i]
                    a2 = diff_ele[j]
                    if not isinstance(a1,float) and not isinstance(a2,float):
                        if a1 not in cty_id:
                            ct += 1
                            cty_id[a1] = ct
                        key1 = ct
                        if a2 not in cty_id:
                            ct += 1
                            cty_id[a2] = ct
                        key2 = ct
                        if (a1,a2) not in cty_tp_val:
                            cty_tp_val[(a1,a2)] = 1
                        else:
                            cty_tp_val[(a1,a2)] += 1
#                         fout.write(a1 + '\t' + a2 + '\t' +'1' +'\n')
    new_tp_val = {}
    for tp in cty_tp_val:
        if cty_tp_val[tp] > 1:
            new_tp_val[tp] = cty_tp_val[tp]

    ct = -1
    cty_id = {}
    for tp in new_tp_val:
        if new_tp_val[tp] > 1:
            if tp[0] not in cty_id:
                ct += 1
                cty_id[tp[0]] = ct  
            if tp[1] not in cty_id:
                ct += 1
                cty_id[tp[1]] = ct

    with open('author_graph_2015.txt', 'w') as fout:
        fout.write('source' + '\t' + 'target' + '\t' + 'value' + '\n')
        for ele in new_tp_val:
            c1 = ele[0]
            c2 = ele[1]
            fout.write(str(cty_id[c1]) + '\t' + str(cty_id[c2]) + '\t' + str(new_tp_val[ele]) + '\n')

    with open('author_graph_id_2015.txt', 'w') as fout:
        fout.write('name' + '\t' + 'group' + '\n')
        for key in cty_id:
            fout.write(key + '\t' + str(cty_id[key]) + '\n')
    return new_tp_val
```
